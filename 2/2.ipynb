This model uses C* of -1 and r0 of 1, providing a logistic function of -ln(x) for range 1 through 10.

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from numpy import *

def cfit(x, a, c, d):
    return c*log(abs((x-a)/(x+a)))+d

domain = arange(.01, .29, .01)
p = [10, 8.57, 7.67, 7.22, 6.88, 6.51, 6.25, 6.02, 5.83, 5.61, 5.42, 5.26, 5.07, 4.83, 4.65, 4.36, 4.29, 4.08, 3.86, 3.63, 3.46, 3.15, 2.95, 2.7, 2.31, 1.9, 1.32, 0]
plt.figure()
plt.scatter(domain, p, label="Raw Data")
popt, pcov = curve_fit(cfit, domain, p)
printer = str(popt)[1:-1].split()
plt.plot(domain, cfit(domain, *popt), label="a={:.5} c={:.5} d={:.5}".format(printer[0], printer[1], printer[2]))
plt.title("Distance vs. Potential")
plt.xlabel("x (m)")
plt.ylabel("Potential (V)")
plt.legend()
plt.show()
```

We find a C value of 18.61 and a D value of 8.579. The optimized curve fit works well, following the general curve shape throughout the middle of the function but struggling to aquire each of the concavities observed at the begining and end of the function.
