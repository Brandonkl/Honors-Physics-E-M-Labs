```python
import numpy as np
import scipy.integrate as integrate
%matplotlib notebook  
import matplotlib
import matplotlib.pyplot as plt
l = np.arange(-1, 1, .01)
j = sum([i**3+1 for i in l])
print("sum of observed points: {:.3f}".format(j))
print("approx integral: {:.3f}".format(.01*j))
c = integrate.quad(lambda X: X**3+1, -1, 1)
print("exact integral: {:.3f}".format(c[0]-c[1]))
f = l**3 + 1 
plt.figure()
plt.plot(l,f)
plt.title("x^3 + 1")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
# Load the standard set of libraries for basic numerical computing.
import numpy as np
# The percent sign is a special character that identifies special functions in Jupyter. This does not work in regular Python.
%matplotlib notebook  
import matplotlib
import matplotlib.pyplot as plt  
import math as m
v = [1.001, 0.0986, 1.002, 0.992, 0.864, 0.608, 0.353]
f = [100, 500, 1000, 2000, 5000, 10000, 20000]
f = [m.log(n) for n in f]
plt.figure()
plt.plot(f, v)
plt.title("Plot of RMS voltage vs ln(frequency)")
plt.xlabel("ln(frequency)")
plt.ylabel("RMS Voltage")
plt.show()
```
