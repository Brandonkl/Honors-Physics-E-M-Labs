```python
# Load the standard set of libraries for basic numerical computing.
import numpy as np
# The percent sign is a special character that identifies special functions in Jupyter. This does not work in regular Python.
%matplotlib notebook  
import matplotlib
import matplotlib.pyplot as plt  
import math as m
v = [1.001, 0.0986, 1.002, 0.992, 0.864, 0.608, 0.353]
f = [.001998, -.007025, .003, -.006018, -.15199, -.6125, -1.067]
plt.figure()
plt.plot(f, v)
plt.title("Plot of RMS voltage vs ln(frequency)")
plt.xlabel("ln(frequency)")
plt.ylabel("RMS Voltage")
plt.show()
```
